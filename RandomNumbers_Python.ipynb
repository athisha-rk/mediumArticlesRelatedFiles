# **Random Numbers in Python**

Computers generated pesudo random numbers. Pesudo Random Number Generators(PRNGs) are based on computational algorithms that can produce apparently random results. Python, like any other programming language uses PRNGs. Python's random generation is based upon Mersenne Twister algorithm that produces 53-bit precision floats.

### **Required Libraries**

```
import numpy as np
```

### **A Quick Example...**

```
import numpy as np
np.random.randint(10,50)
```

```
np.random.random()
```

### **np.random.randint()**

```
#Generate a random integer between 10 to 80
#low = 10 (inclusive); high = 80 (exclusive)
print(np.random.randint(10,80))

print(np.random.randint(1999,2022))

#Generate random integers between 10 to 80
#2D Array of size 2x3
print(np.random.randint(10,80,(2,3)))

#1D Array of size 10
print(np.random.randint(10,80,10))
```

### **np.random.random()**

```
#Generate a random float in the interval [0.0,1.0)
print(np.random.random())

#Generate random floats in the interval [0.0,1.0) 
#1D Array of size 5
print(np.random.random(5))

#2D array of size 2x2
print(np.random.random((2,2)))
```

### **np.random.get_state()**

```
#Get the internal state of the generator
np.random.get_state()
```

### **np.random.seed()**

```
#Seed The Generator
np.random.seed(10)
```

```
np.random.rand()
#The generator creates a random number based on the seed value, 
#so if the seed value is 10, you will always get 0.771320643266746 as 
#the first number
```
